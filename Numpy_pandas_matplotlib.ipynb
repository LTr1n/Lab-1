### 1.1 NUMPY

```python
import numpy as np

# tạo một mảng 1 chiều
mang_1D = np.array([1,2,3,4,5,6,7,8])
print(mang_1D)

# tạo một mảng 2 chiều
mang_2D = np.array([[1,3,5,7],[2,4,6,8]])
print(mang_2D)

# tạo mảng ngẫu nhiên
mang_nD = np.random.rand(4,4)
print(mang_nD)
```

CÁC THAO TÁC CƠ BẢN TRONG NUMPY

```python
# Tạo một mảng
mang_1D = np.array([1,2,3,4,5,6,7,8])
print(mang_1D)

# cộng và nhân mảng với một số
print(mang_1D + 5)
print(mang_1D * 5)

# tính tổng , trung bình , lớn nhất , nhỏ nhất
print(mang_1D.sum())
print(mang_1D.mean())
print(mang_1D.max())
print(mang_1D.min())

#Lấy phần tử
print(mang_1D[2])
```

PANDAS

### 2.1 Tạo DataFrame

```python
import pandas as pd

# Tạo dataframe và dictionary
data = {
    "tên" : ["Anh" , "yêu" , "em"],
    "Tuổi": [18, 20 , 16],
    "diem": [3 , 7 , 10]
}
df = pd.DataFrame(data)
print(df)
```

### 2.2 Thao tác cơ bản

```python
print(df["Tuổi"])

# lấy hàng thứ 2

print(df.iloc[1])

# Lấy ra người có tuổi lớn hơn 18
print(df[df["Tuổi"]>18])

# tính điểm trung bình của những người đó
print(df["diem"].mean())
```

3.1 Biểu đồ đường

```python
import matplotlib.pyplot as plt

x = [5, 8, 7, 4, 32, 99]
y = [20, 87, 64, 55, 30, 100]

plt.plot(x, y, label="Biểu đồ đường", color="pink", marker="o")
plt.title("Biểu đồ đường")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.legend()
plt.show()

```

### Biểu đồ cột

```python
tên = ["Hải" , "Việt" , "Cường" , "Thiện" , "Trọng"]
điểm = [2 , 2.5 , 1 , 1.5 , 0.5]

plt.bar(tên , điểm , color = "red")
plt.title("Biểu đồ cột")
plt.xlabel("Tên")
plt.ylabel("điểm")
plt.show()
```

### 3.3 Biểu đồ phân tán

```python
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.rand(100) * 100

plt.scatter(x , y , s = sizes , alpha = 0.5 , c = "green")
plt.title("Biểu đồ phân tán")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.show() 
```
